## ğŸ”§ 3. ComfyUI Installation starten

**Das ist der Hauptinstallationsschritt - kann 15-20 Minuten dauern!**

```python
# ComfyUI Installation starten
import subprocess
import os

print("ğŸš€ COMFYUI INSTALLATION")
print("=======================")
print("â±ï¸  Dies kann 15-20 Minuten dauern...")
print("ğŸ“Š Live-Output wird angezeigt...")
print()

# Sicherstellen, dass wir im richtigen Verzeichnis sind
if not os.path.exists('/comfyui-paperspace-notebook/run.sh'):
    print("âŒ run.sh nicht gefunden! Stelle sicher, dass das Repository geklont wurde.")
else:
    os.chdir('/comfyui-paperspace-notebook')
    
    # Script ausfÃ¼hrbar machen
    subprocess.run(['chmod', '+x', 'run.sh'])
    
    # Installation starten mit Live-Output
    print("ğŸ”„ Starte Installation...")
    print("=" * 50)
    
    process = subprocess.Popen(['./run.sh'], 
                              stdout=subprocess.PIPE, 
                              stderr=subprocess.STDOUT,
                              universal_newlines=True,
                              bufsize=1)
    
    # Live-Output anzeigen
    for line in process.stdout:
        print(line.rstrip())
    
    # Warten bis Prozess beendet ist
    process.wait()
    
    if process.returncode == 0:
        print()
        print("=" * 50)
        print("ğŸ‰ INSTALLATION ERFOLGREICH ABGESCHLOSSEN!")
        print("=" * 50)
    else:
        print()
        print("=" * 50)
        print(f"âŒ Installation fehlgeschlagen (Exit Code: {process.returncode})")
        print("=" * 50)
```

## ğŸ” 5. Installation Ã¼berprÃ¼fen

```python
# Installation Status prÃ¼fen
import subprocess
import os
import requests
from pathlib import Path

print("ğŸ” INSTALLATION STATUS")
print("======================")

# 1. PrÃ¼fe ob ComfyUI-Verzeichnis existiert
comfyui_path = Path('/comfyui-paperspace-notebook/ComfyUI')
if comfyui_path.exists():
    print("âœ… ComfyUI-Verzeichnis gefunden")
else:
    print("âŒ ComfyUI-Verzeichnis nicht gefunden")

# 2. PrÃ¼fe Virtual Environment
venv_path = Path('/comfyui-paperspace-notebook/venv')
if venv_path.exists():
    print("âœ… Virtual Environment erstellt")
else:
    print("âŒ Virtual Environment nicht gefunden")

# 3. PrÃ¼fe laufende Python-Prozesse
try:
    ps_result = subprocess.run(['ps', 'aux'], capture_output=True, text=True)
    comfyui_processes = [line for line in ps_result.stdout.split('\n') if 'main.py' in line and 'ComfyUI' in line]
    
    if comfyui_processes:
        print(f"âœ… ComfyUI-Prozess lÃ¤uft: {len(comfyui_processes)} Prozess(e)")
        for proc in comfyui_processes[:2]:  # Zeige max. 2 Prozesse
            print(f"   ğŸ“Š {proc.strip()}")
    else:
        print("âŒ Kein ComfyUI-Prozess gefunden")
except:
    print("âš ï¸  Prozess-Check fehlgeschlagen")

# 4. PrÃ¼fe Port 6006
try:
    response = requests.get('http://localhost:6006', timeout=5)
    if response.status_code == 200:
        print("âœ… ComfyUI antwortet auf Port 6006")
    else:
        print(f"âš ï¸  Port 6006 antwortet mit Status: {response.status_code}")
except requests.exceptions.ConnectionError:
    print("âŒ ComfyUI nicht erreichbar auf Port 6006")
except requests.exceptions.Timeout:
    print("â±ï¸  Timeout bei Port 6006 - ComfyUI startet mÃ¶glicherweise noch")
except Exception as e:
    print(f"âš ï¸  Port-Check Fehler: {e}")

# 5. Modelle prÃ¼fen
models_path = Path('/storage/ComfyUI/models')
if models_path.exists():
    print("âœ… Modelle-Verzeichnis gefunden")
    
    # ZÃ¤hle Modelle in verschiedenen Kategorien
    checkpoints = list((models_path / 'checkpoints').glob('*.safetensors')) if (models_path / 'checkpoints').exists() else []
    vae = list((models_path / 'vae').glob('*.safetensors')) if (models_path / 'vae').exists() else []
    
    print(f"   ğŸ“¦ Checkpoints: {len(checkpoints)} Modelle")
    print(f"   ğŸ¨ VAE: {len(vae)} Modelle")
else:
    print("âŒ Modelle-Verzeichnis nicht gefunden")

print()
print("ğŸš€ NÃ„CHSTE SCHRITTE:")
print("- Ã–ffne die ComfyUI-URL in einem neuen Tab")
print("- Bei 'Bad Gateway': Warte 2-3 Minuten und versuche erneut")
print("- PrÃ¼fe die deutsche Dokumentation in docs/ANLEITUNG_DEUTSCH.md")
```
