## ğŸ”§ 3. ComfyUI Installation starten

**Das ist der Hauptinstallationsschritt - kann 15-20 Minuten dauern!**

```python
# Optional: Nur fehlende Dependencies installieren (schneller)
import subprocess
import os

print("ğŸ”§ QUICK FIX: Nur fehlende Dependencies installieren")
print("====================================================")

if os.path.exists('/comfyui-paperspace-notebook/scripts/fix_dependencies.sh'):
    os.chdir('/comfyui-paperspace-notebook')
    
    print("ğŸ”„ Installiere fehlende Pakete (torchsde, torchaudio, scipy, einops)...")
    
    # Fix-Script ausfÃ¼hren
    subprocess.run(['chmod', '+x', 'scripts/fix_dependencies.sh'])
    result = subprocess.run(['./scripts/fix_dependencies.sh'], capture_output=True, text=True)
    
    print(result.stdout)
    if result.returncode == 0:
        print("âœ… Alle fehlenden Dependencies installiert!")
    else:
        print("âš ï¸ Einige Dependencies konnten nicht installiert werden")
        print(result.stderr)
else:
    print("âš ï¸ Fix-Script nicht gefunden. FÃ¼hre vollstÃ¤ndige Installation aus.")
```

```python
# ComfyUI Installation starten mit verbesserter Fehlerbehandlung
import subprocess
import os

print("ğŸš€ COMFYUI INSTALLATION (Mit Robustheit-Verbesserungen)")
print("=========================================================")
print("â±ï¸  Dies kann 15-20 Minuten dauern...")
print("ğŸ“Š Live-Output wird angezeigt...")
print()

# Sicherstellen, dass wir im richtigen Verzeichnis sind
if not os.path.exists('/comfyui-paperspace-notebook/run.sh'):
    print("âŒ run.sh nicht gefunden! Stelle sicher, dass das Repository geklont wurde.")
else:
    os.chdir('/comfyui-paperspace-notebook')
    
    # Zuerst Repository aktualisieren fÃ¼r neueste Fixes
    print("ğŸ”„ Aktualisiere Repository fÃ¼r neueste Fixes...")
    subprocess.run(['git', 'pull', 'origin', 'master'])
    
    # Scripts ausfÃ¼hrbar machen
    subprocess.run(['chmod', '+x', 'run.sh'])
    subprocess.run(['chmod', '+x', 'scripts/robust_installer.sh'])
    subprocess.run(['chmod', '+x', 'scripts/fix_dependencies.sh'])
    
    # Robuste Installation verwenden (mit Fallback)
    print("ğŸ”„ Starte robuste Installation...")
    print("=" * 50)
    
    # Versuche zuerst die robuste Version
    if os.path.exists('scripts/robust_installer.sh'):
        print("âœ… Verwende neue robuste Installation mit Retry-Mechanismen...")
        process = subprocess.Popen(['./scripts/robust_installer.sh'], 
                                  stdout=subprocess.PIPE, 
                                  stderr=subprocess.STDOUT,
                                  universal_newlines=True,
                                  bufsize=1)
    else:
        print("âš ï¸ Fallback auf Standard-Installation...")
        process = subprocess.Popen(['./run.sh'], 
                                  stdout=subprocess.PIPE, 
                                  stderr=subprocess.STDOUT,
                                  universal_newlines=True,
                                  bufsize=1)
    
    # Live-Output anzeigen
    for line in process.stdout:
        print(line.rstrip())
    
    # Warten bis Prozess beendet ist
    process.wait()
    
    if process.returncode == 0:
        print()
        print("=" * 50)
        print("ğŸ‰ INSTALLATION ERFOLGREICH ABGESCHLOSSEN!")
        print("=" * 50)
    else:
        print()
        print("=" * 50)
        print(f"âš ï¸ Installation mit Warnungen beendet (Exit Code: {process.returncode})")
        print("ğŸ”§ Versuche Reparatur mit fix_dependencies.sh...")
        print("=" * 50)
        
        # Versuche Dependencies zu reparieren
        if os.path.exists('scripts/fix_dependencies.sh'):
            fix_process = subprocess.Popen(['./scripts/fix_dependencies.sh'],
                                          stdout=subprocess.PIPE,
                                          stderr=subprocess.STDOUT,
                                          universal_newlines=True,
                                          bufsize=1)
            for line in fix_process.stdout:
                print(line.rstrip())
            fix_process.wait()
            
            if fix_process.returncode == 0:
                print("âœ… Dependencies erfolgreich repariert!")
            else:
                print("âš ï¸ Manuelle Intervention erforderlich")
```
